```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import pandas as pd
import matplotlib
matplotlib.rcParams['figure.figsize'] = [20, 20]
matplotlib.rcParams['figure.dpi'] = 600

df = pd.read_csv('/home/suhail/Downloads/su_lat_lon.csv')

lat = df['a_obs.latitude1'].values
lon = df['a_obs.longitude1'].values 

fig = plt.figure(figsize=(14, 12))

m = Basemap(llcrnrlon=-84.0, llcrnrlat=32.0, urcrnrlon=-71.4, urcrnrlat=45.25,
projection='lcc',resolution='h',width=10E6, height=8E6, lat_0=39, lon_0=-75.)
# m = Basemap(projection='lcc', resolution='h',
#             width=10E6, height=8E6, 
#             lat_0=40, lon_0=-75,)
#m.scatter(lon, lat, latlon=True,marker='o', c='red')

#m.fillcontinents(lake_color='#DDEEFF')
m.drawcoastlines()
m.drawstates()
#m.drawlsmask()
m.shadedrelief(scale=2)
pts = m.scatter(lon, lat, latlon=True,marker='o', c='red',zorder=2)
m.fillcontinents(lake_color='#DDEEFF',zorder=1)
# Map (long, lat) to (x, y) for plotting
# x, y = m(-75.24, 39.88)
# plt.plot(x, y, 'ok', markersize=5)
# plt.text(x, y, ' Philadephia', fontsize=12);

# cfig = plt.gcf()
# cfig.set_size_inches(12.460*.90,9.520*.90, forward=True)
plt.legend([pts],['Station'],frameon=True)

plt.savefig('/home/suhail/Downloads/map_1.png')
    
```
